### Packages and Environment setting

```python
## Packages
import numpy as np
import pandas as pd
import time
import datetime
pd.set_option('display.max_columns', 300)
pd.set_option('display.max_rows', 1000)
pd.options.display.max_colwidth = 1000
import os
import math

## env
PATH_RAW = '/Users/jerry/Desktop/example/'        ## input path
PATH_OUT = '/Users/jerry/Desktop/dataBase/DataProcessed/' ## output path
encoding = 'utf-8'
```

### Header columns information

```python
os.chdir(PATH_RAW)
lst = os.listdir()
lst = [lst for lst in lst if '.csv' in lst]

data_store = pd.DataFrame(columns = ['index', 'col'])

for x in list(range(len(lst))):
    temp = pd.read_csv(str(lst[x]), nrows = 0, encoding = encoding)
    temp = temp.T.reset_index().reset_index()
    temp.columns = ['index', 'col']
    temp['index'] = str(lst[x])
    data_store = pd.concat([data_store, temp])
    
data_store['index'] = data_store['index'].str.replace('.csv', '')

data_store = data_store.sort_values(by = 'index')
data_store.columns = ['csv_name', 'col_name']

os.chdir(PATH_OUT)
data_store.to_csv('header_all.csv')
display(data_store)
```

### With dtypes information

```python
os.chdir(PATH_RAW)
lst = os.listdir()
lst = [lst for lst in lst if '.csv' in lst]
    
data_store = pd.DataFrame()

for x in list(range(len(lst))):
    temp = pd.read_csv(str(lst[x]), encoding = encoding, error_bad_lines=False, nrows = 30)
    temp = pd.DataFrame(temp.dtypes, columns = ['dtypes']).reset_index()
    temp['csv_name'] = str(lst[x])

    data_store = pd.concat([data_store, temp])

data_store['csv_name'] = data_store['csv_name'].str.replace('.csv', '')
data_store = data_store.sort_values(by = 'csv_name')
data_store = data_store[['csv_name', 'index', 'dtypes']]
data_store = data_store.rename(columns = {'index': 'col_name'})

os.chdir(PATH_OUT)
data_store.to_csv('header_all_dtypes.csv')   

display(data_store)
```
